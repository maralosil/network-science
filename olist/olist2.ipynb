```python
import pandas as pd
import networkx as nx
import itertools
import warnings
warnings.filterwarnings("ignore")
```

```python
items = pd.read_csv('datasets/olist_order_items_dataset.csv')
products = pd.read_csv('datasets/olist_products_dataset.csv')
```

```python
items_products = items.merge(products, how = 'left', on = 'product_id')
```

```python
order_id_categories = {}
for index, row in items_products.iterrows():
    if row.order_id not in order_id_categories:
        order_id_categories[row.order_id] = set()
    order_id_categories[row.order_id].add(row.product_category_name)

print(order_id_categories['30bdf3d824d824610a49887486debcaf'])    
```

```python
G = nx.Graph()

for order_id, categories in order_id_categories.items():
    for category1, category2 in itertools.combinations(categories, 2):
        if G.has_edge(category1, category2):
            G[category1][category2]['weight'] += 1
        else:
            G.add_edge(category1, category2, weight=1)
```
